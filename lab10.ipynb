```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
# 2. Displaying two different images
img = cv2.imread('arc.jpeg')
img2 = cv2.imread('triomphe.jpg')
rgb_img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
rgb_img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_img)
plt.show()
plt.imshow(rgb_img2)
plt.show()
```

```python
# 3. Displaying the gradient magnitude
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
sob_x = cv2.Sobel(gray, cv2.CV_64F, 1, 0)
sob_y = cv2.Sobel(gray, cv2.CV_64F, 0, 1)
grad_mag = np.sqrt(sob_x**2 + sob_y**2)
gray2 = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
sob_x2 = cv2.Sobel(gray2, cv2.CV_64F, 1, 0)
sob_y2 = cv2.Sobel(gray2, cv2.CV_64F, 0, 1)
grad_mag2 = np.sqrt(sob_x2**2 + sob_y2**2)
plt.imshow(grad_mag, cmap = 'gray')
plt.show()
plt.imshow(grad_mag2, cmap = 'gray')
plt.show()
```

```python
# 4.  Find and display edges
gray_smooth = cv2.GaussianBlur(gray, (3,3), 0)
dst = cv2.Canny(gray_smooth, 55, 200)
edge_thresh = dst > 0
img_disp = img.copy()
img_disp[edge_thresh] = (0,255,255)
gray_smooth2 = cv2.GaussianBlur(gray2, (3,3), 0)
dst2 = cv2.Canny(gray_smooth2, 30, 120)
edge_thresh2 = dst2 > 0
img_disp2 = img2.copy()
img_disp2[edge_thresh2] = (0,255,255)
rgb_disp = cv2.cvtColor(img_disp, cv2.COLOR_BGR2RGB)
rgb_disp2 = cv2.cvtColor(img_disp2, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_disp)
plt.show()
plt.imshow(rgb_disp2)
plt.show()
```

```python
# 5. Find and display Harris Corners
block_size = 4
kernel_size = 17
k = 0.04
dst = cv2.cornerHarris(gray, block_size, kernel_size, k)
thresh_val = 0.03 * dst.max()
corner_thresh = dst > thresh_val
img_dis = img.copy()
img_dis[corner_thresh] = (0, 255, 0)
block_size2 = 10
kernel_size2 = 15
k2 = 0.04
dst2 = cv2.cornerHarris(gray2, block_size2, kernel_size2, k2)
thresh_val2 = 0.03 * dst2.max()
corner_thresh2 = dst2 > thresh_val2
img_dis2 = img2.copy()
img_dis2[corner_thresh2] = (0, 255, 0)
rgb_dis = cv2.cvtColor(img_dis, cv2.COLOR_BGR2RGB)
rgb_dis2 = cv2.cvtColor(img_dis2, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_dis)
plt.show()
plt.imshow(rgb_dis2)
plt.show()
```

```python
# Find and display ORB Features
orb = cv2.ORB_create()
keypoints = orb.detect(img)
img_display = img.copy()
img_display = cv2.drawKeypoints(img_display, keypoints, None, (255,0,255))
orb2 = cv2.ORB_create()
keypoints2 = orb2.detect(img2)
img_display2 = img2.copy()
img_display2 = cv2.drawKeypoints(img_display2, keypoints2, None, (255,0,255))
rgb_display = cv2.cvtColor(img_display, cv2.COLOR_BGR2RGB)
rgb_display2 = cv2.cvtColor(img_display2, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_display)
plt.show()
plt.imshow(rgb_display2)
plt.show()
```
