```python
# libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
# 2 add noise method
def add_noise(img, strength):
  img = img.astype(float)
  noise = np.random.randint(-strength, strength, img.shape)
  result = img + noise
  result = np.clip(result, 0, 255)
  result = result.astype(np.uint8)
  return result

```

```python
# 3 Display original, little noise, more noise
image = cv2.imread('golf.jpg')
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
little_noise = add_noise(gray_image, 200)
more_noise = add_noise(gray_image, 400)
plt.imshow(gray_image, cmap='gray')
plt.title('Original')
plt.show()
plt.imshow(little_noise, cmap='gray')
plt.title('Little noise')
plt.show()
plt.imshow(more_noise, cmap='gray')
plt.title('More noise')
plt.show()
```

```python
# 4. Denoising 
median_filter = cv2.medianBlur(more_noise, 13)
gaussian = cv2.GaussianBlur(more_noise, (27, 27), 0)
blur = cv2.blur(more_noise, (22,22))
plt.imshow(median_filter, cmap='gray')
plt.title('Median filter')
plt.show()
plt.imshow(gaussian, cmap='gray')
plt.title('Gaussian blurring')
plt.show()
plt.imshow(blur, cmap = 'gray')
plt.title('Blurring')
plt.show()
```

5. We can evaluate the effectiveness of denoising methods numerically using a metric like sum of squared differences. The method with a lower a sum of squared difference would be quantitatively more efficient. This idea would be used for an image where you added noise yourself because you as the programmer know how much noise you added. Therefore, you can accurately compare the difference between the noise reduced from a method, by calculating a difference between the noisy image pixels and pixels after the denoising method implementation, and the noise you added to the original image. Using sum of squared differences, if noise reduced and noise added are equal, then that yields 0,or the original image, and would indicate a great denoising method. Using an already noisy image would make it hard to evaluate sum of squared differences because there is no reference, to compare the denoising method effects to, to determine effectiveness.
