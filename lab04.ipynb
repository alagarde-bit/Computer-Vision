```python
# importing libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 2. Separate color channels
img = cv2.imread("my_image.jpg")
rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(rgb)
plt.title("Original Image")
plt.show()

b, g, r = cv2.split(img)
plt.imshow(b, cmap='gray', vmin=0, vmax=255)
plt.title('Blue Channel')
plt.show()
plt.imshow(g, cmap='gray', vmin=0, vmax=255)
plt.title('Green Channel')
plt.show()
plt.imshow(r, cmap='gray', vmin=0, vmax=255)
plt.title('Red Channel')
plt.show()

new_img = cv2.merge((b,r,g))
plt.imshow(new_img)
plt.title('Recombined Wrong Order')
plt.show()

r2, c = g.shape[:2]
g_center =  ((c - 1) / 2, (r2 - 1) / 2)
T = cv2.getRotationMatrix2D(g_center, 30, 1.0)
g_rot = cv2.warpAffine(g,T,(c,r2))
transformed_img = cv2.merge((b,g_rot,r))
plt.imshow(transformed_img)
plt.title("Transformed Image")
plt.show()
```

```python
# 3. Grayscale
img2 = cv2.imread('Alex_Lagarde_Profile.jpg')
gray = cv2.cvtColor(img2, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
plt.title("Gray Image")
plt.show()
```

```python
# convert gray function
def convert_gray(image):
  img = image
  i = cv2.imread(img)
  i = i.astype(np.float)
  B, G, R = cv2.split(i)
  new_B = 0.07 * B
  new_G = 0.72 * G
  new_R = 0.21 * R
  total = new_B + new_G + new_R
  total = np.clip(total, 0, 255)
  total = total.astype(np.uint8)
  return total

disp = convert_gray('Alex_Lagarde_Profile.jpg')
plt.imshow(disp, cmap='gray')
plt.plot()
```
