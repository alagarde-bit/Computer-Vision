```python
# import libraries
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('waldo_scene1.jpg')
template = cv2.imread('waldo.jpg')
w,h = template.shape[:2]
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray_template = cv2.cvtColor(template, cv2.COLOR_BGR2GRAY)
method = cv2.TM_CCORR_NORMED
res = cv2.matchTemplate(gray_img,gray_template, method)
plt.imshow(res, cmap='gray')
plt.title('Normalized Cross Correlation')
plt.show()
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(img,top_left, bottom_right, (0,255.0), 3)
plt.imshow(img,cmap = 'gray')
plt.title('Normalized Cross Correlation Rectangle')
plt.show()

```

```python
method2 = cv2.TM_CCORR
res2 = cv2.matchTemplate(gray_img,gray_template, method2)
plt.imshow(res2, cmap='gray')
plt.title('Non-Normalized Cross Correlation')
plt.show()
min_val2, max_val2, min_loc2, max_loc2 = cv2.minMaxLoc(res2)
top_left2 = max_loc2
bottom_right2 = (top_left2[0] + w, top_left2[1] + h)
cv2.rectangle(img,top_left2, bottom_right2, (0,255,0), 3)
plt.imshow(img,cmap = 'gray')
plt.title('Non-Normalized Cross Correlation Rectangle')
plt.show()
```

Although both methods found waldo, the non-normalized cross correlation method added another rectangle on the opposite side of the image from where Waldo was found, while Normalized cross correlation yielded one rectangle, which correctly found Waldo.
